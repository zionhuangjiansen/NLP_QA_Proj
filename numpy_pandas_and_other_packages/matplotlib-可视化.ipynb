```python
### pip install matplot lib
import matplotlib.pyplot as plt
import numpy as np
plt.plot(np.arange(10))
plt.figure()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
from numpy.random import randn
plt.plot(randn(50).cumsum(), 'k--') # 灰色
ax1.hist(randn(100), bins=20, color='k', alpha=0.36)
ax2.scatter(np.arange(30), np.arange(30)+3*randn(30))
```

```python
plt.plot(randn(50).cumsum(), 'g--')
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y,linestyle='--',color='#CECECE')
```

```python
### Series 画图
from pandas import Series, DataFrame
s = Series(randn(10).cumsum(), index=np.arange(0,100,10))
s.plot()
```

```python
df = DataFrame(np.random.randn(10, 4).cumsum(0). columns=['A','B','C','D'], index=np.range(0,100,10))
df.plot()
```
